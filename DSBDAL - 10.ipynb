#### **0. Loading the Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

#### **2. Create a histogram for each feature in the dataset to illustrate the feature distributions.**

```python
fig, ax = plt.subplots(nrows=4, ncols=3, figsize=(10,8))
plt.subplots_adjust(wspace=0.2, hspace=0.4)

cols = iris.columns[:-1]

for i in range(4):
    sns.histplot(ax = ax[i,0], x = cols[i], data = iris.loc[iris['species']=='setosa'], color='palegreen')
    sns.histplot(ax = ax[i,1], x = cols[i], data = iris.loc[iris['species']=='versicolor'], color='cyan')
    sns.histplot(ax = ax[i,2], x = cols[i], data = iris.loc[iris['species']=='virginica'], color='pink')

for axis in ax.flatten():
  axis.set_xlabel(' '); axis.set_ylabel(' ');

cols = ['Species = {}'.format(col) for col in ['Setosa', 'Versicolor', 'Virginica']]
rows = ['{}'.format(row) for row in ['sepal_width', 'sepal_length', 'petal_width', 'petal_length']]

for axis, col in zip(ax[0], cols):
    axis.set_title(col)
for axis, row in zip(ax[:,0], rows):
    axis.set_ylabel(row, rotation=90, size='large');
```

###### 1. Sepal width is more granular in case of setosa species and least in case of versicolor.
###### 2. Sepal length, petal width and length are ordered : setosa < versicolor < virginica.
###### 3. The petal length distribution in setosa closely resembles a normal distribution.

#### **3. Create a boxplot for each feature in the dataset.**

```python
fig, ax = plt.subplots(nrows=4, ncols=3, figsize=(10,8))
plt.subplots_adjust(wspace=0.2, hspace=0.4)

cols = iris.columns[:-1]

for i in range(4):
    sns.boxplot(ax = ax[i,0], y = cols[i], data = iris.loc[iris['species']=='setosa'], color='palegreen')
    sns.boxplot(ax = ax[i,1], y = cols[i], data = iris.loc[iris['species']=='versicolor'], color='cyan')
    sns.boxplot(ax = ax[i,2], y = cols[i], data = iris.loc[iris['species']=='virginica'], color='pink')

for axis in ax.flatten():
  axis.set_xlabel(' '); axis.set_ylabel(' ');

cols = ['Species = {}'.format(col) for col in ['Setosa', 'Versicolor', 'Virginica']]
rows = ['{}'.format(row) for row in ['sepal_width', 'sepal_length', 'petal_width', 'petal_length']]

for axis, col in zip(ax[0], cols):
    axis.set_title(col)
for axis, row in zip(ax[:,0], rows):
    axis.set_ylabel(row, rotation=90, size='large');
```

#### **4. Compare distributions and identify outliers**

###### 1. The mean for sepal width is similar for species setosa and virginica (and both have 2 outlying values), lower for versicolor.
###### 2. The mean sepal length for setosa < versicolor < virginica (has 1 outlier).
###### 3. Petal width and the range of distribution of values for setosa (2 outliers) < versicolor < virginica.
###### 4. Similar ordering is observed in case of petal length. Setosa has 3 outliers and versicolor has 1.
