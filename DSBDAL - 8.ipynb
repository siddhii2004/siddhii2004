# Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.shape
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.isna().sum()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True);
```

```python
titanic['deck'].fillna(titanic['deck'].mode().iloc[0], inplace=True);
```

```python
titanic.isna().sum().sum()
```

```python
titanic.dropna(inplace=True)
```

```python
titanic.isna().sum().sum()
```

# Visualize Patterns using Seaborn

```python
sns.set_style('darkgrid')
sns.pairplot(hue='survived', data=titanic, vars=['pclass','adult_male','age','alone'], palette=['red','green']);
```

##### 1. From the above plot, we can infer that maximum deaths were from people with class=3, minimum for class=1.
##### 2. Most of the people who died were adult males.
##### 3. The ages of most passengers were between 10-40.
##### 4. People who were alone, survived less in number.

```python
sns.countplot(x='sex', hue='survived', data=titanic, palette=['red','green']);
```

##### More females survived as compared to males.

##### **2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.**


```python
sns.set(rc={'figure.figsize':(15,5)})
sns.histplot(x='fare', data=titanic, color='blue')
plt.xlabel('Fare')
plt.title('Fare Histogram', size='large');
```

##### The average ticket fare lies between 0-100 units. There are outlying values too where some passengers paid > 500.
